## Duration

```python
df['Duration'].unique()
```

```python
durations = list(df['Duration'])
durations
```

```python
for i in range(len(durations)):
    durations[i] = durations[i].split()
    if(len(durations[i]) == 1):
        durations[i].append('0m')
durations
```

```python
hours= []
minutes=[]
for duration in durations:
    hours.append(duration[0])
    minutes.append(duration[1])
```

```python
df['Duration_hour'] = hours
df['Duration_minute'] = minutes
```

```python
df.head()
```

```python
# df[df['Duration_hour'].str.contains('m')]
```

```python
df['Duration_hour'] = df['Duration_hour'].str.replace('h','')
df['Duration_minute'] = df['Duration_minute'].str.replace('m','')
```

```python
df.head()
```

```python
df['Duration_hour'] = df['Duration_hour'].astype('int')
df['Duration_minute'] = df['Duration_minute'].astype('int')
```

```python
df.drop(columns=['Duration'],inplace=True)
```
